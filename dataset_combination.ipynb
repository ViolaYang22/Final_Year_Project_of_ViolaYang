```python
import pandas as pd
```

```python
dataset_2019 = pd.read_csv('combined_dataset_2019.csv', sep=',')
dataset_2019
```

```python
dataset_2020 = pd.read_csv('combined_dataset_2020.csv', sep=',')
dataset_2020
```

```python
dataset_2021 = pd.read_csv('combined_dataset_2021.csv', sep=',')
dataset_2021
```

```python
#Remove the feature NPI
dataset_2020.drop(columns='NPI', inplace=True)
dataset_2019.drop(columns='NPI', inplace=True)
```

```python
#combine the dataset of 2019 & 2020
dataset_of_1920_combination = pd.concat([dataset_2019,dataset_2020], ignore_index = True, sort = False)
dataset_of_1920_combination
```

```python
#add zero to the merged dataset
dataset_of_1920_combination.fillna(0, inplace = True)
dataset_of_1920_combination
```

```python
#store this two datasets and use as training and testing sets respectively
dataset_of_1920_combination.to_csv('dataset_of_1920_combination.csv', index = False)
dataset_2021.to_csv('combined_dataset_2021.csv', index = False)
```
