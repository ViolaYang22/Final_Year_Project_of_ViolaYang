```python
import pandas as pd
```

```python
#import LEIE dataset
LEIE_col_names =['LASTNAME','FIRSTNAME','MIDNAME','BUSNAME','GENERAL','SPECIALTY','UPIN','NPI','DOB','ADDRESS','CITY','STATE','ZIP','EXCLTYPE','EXCLDATE','REINDATE','WAIVERDATE','WVRSTATE']
LEIE_data = pd.read_csv('LEIE Dataset.csv', header = None, names = LEIE_col_names)
LEIE_data = pd.DataFrame(LEIE_data).loc[1:].reset_index(drop=True)
LEIE_data
```

```python
#Extract data on frauds committed in 2019 and later
# Convert 'EXCLDATE' column to datetime for easier date comparison
LEIE_data['EXCLDATE'] = pd.to_datetime(LEIE_data['EXCLDATE'], errors='coerce')

# Define the threshold date for fraud detection (January 1, 2019)
threshold_date = pd.Timestamp('2019-01-01')

# Filter the data to extract frauds committed in 2019 and later
fraud_data_2019 = LEIE_data[(LEIE_data['EXCLDATE'] >= threshold_date) & (LEIE_data['NPI'] != '0000000000')]

# Reset the index of the resulting DataFrame
fraud_data_2019.reset_index(drop=True, inplace=True)
```

```python
fraud_data_2019
```

```python
#Save the generated fraud data in 2019
fraud_data_2019.to_csv('fraud_Data_2019.csv',index = False)
```
