```python
import pandas as pd
from tqdm import tqdm
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
#First of all, import these data for processing
col_names_B = ['Rndrng_NPI', 'Rndrng_Prvdr_Gndr', 'Rndrng_Prvdr_Type','Rndrng_Prvdr_State_FIPS', 'Tot_HCPCS_Cds','Tot_Benes', 'Tot_Srvcs', 'Tot_Sbmtd_Chrg', 'Tot_Mdcr_Pymt_Amt', 'Bene_Avg_Risk_Scre']
col_names_D = ['PRSCRBR_NPI', 'Prscrbr_Gndr','Prscrbr_Type', 'Prscrbr_State_FIPS','Tot_Clms', 'Tot_30day_Fills', 'Tot_Day_Suply', 'Tot_Drug_Cst', 'Tot_Benes']
col_names_DMEPOS = ['Rfrg_NPI', 'Rfrg_Prvdr_Gndr', 'Rfrg_Prvdr_Type', 'Rfrg_Prvdr_State_FIPS','Tot_Suplrs', 'Tot_Suplr_HCPCS_Cds','Tot_Suplr_Benes', 'Tot_Suplr_Clms', 'Tot_Suplr_Srvcs', 'Suplr_Sbmtd_Chrgs', 'Suplr_Mdcr_Pymt_Amt']

```

```python
#import the part B data of 2021
Part_B_data_2021 = pd.read_csv('Medicare_Physician_Other_Practitioners_by_Provider_2021.csv',sep=',',usecols=col_names_B)
Part_B_data_df_2021 = pd.DataFrame(Part_B_data_2021).iloc[1:].reset_index(drop=True)
Part_B_data_df_2021
```

```python
#import the part B data of 2020
Part_B_data_2020 = pd.read_csv('Medicare_Physician_Other_Practitioners_by_Provider_2020.csv',sep=',',usecols=col_names_B,encoding='ISO-8859-1')
Part_B_data_df_2020 = pd.DataFrame(Part_B_data_2020).iloc[1:].reset_index(drop=True)
Part_B_data_df_2020
```

```python
#import the part B data of 2019 
Part_B_data_2019 = pd.read_csv('Medicare_Physician_Other_Practitioners_by_Provider_2019.csv',sep=',',usecols=col_names_B,encoding='ISO-8859-1')
Part_B_data_df_2019 = pd.DataFrame(Part_B_data_2019).iloc[1:].reset_index(drop=True)
Part_B_data_df_2019
```

```python
#import the list of doctors who involved in fraudulent activities in 2021
fraud_data_df_2021 = pd.read_csv('fraud_Data_2021.csv', index_col=0)
fraud_data_df_2021
```

```python
#import the list of doctors who involved in fraudulent activities in 2019
fraud_data_df_2019 = pd.read_csv('fraud_Data_2019.csv', index_col=0)
fraud_data_df_2019
```

```python
#Generate the LSIT of the NPI in the LEIE dataset of each year
list_of_2021_fraud_data = fraud_data_df_2021['NPI'].values.tolist()
list_of_2020_fraud_data = fraud_data_df_2020['NPI'].values.tolist()
list_of_2019_fraud_data = fraud_data_df_2019['NPI'].values.tolist()
```

```python
#Label Part B dataset of 2021
Part_B_data_df_2021['Exclusion'] = 0

for index, row in tqdm(Part_B_data_df_2021.iterrows()):
    if row['Rndrng_NPI'] in list_of_2021_fraud_data:
        Part_B_data_df_2021.at[index, 'Exclusion'] = 1
```

```python
#Label Part B dataset of 2020
Part_B_data_df_2020['Exclusion'] = 0

for index, row in tqdm(Part_B_data_df_2020.iterrows()):
    if row['Rndrng_NPI'] in list_of_2020_fraud_data:
        Part_B_data_df_2020.at[index, 'Exclusion'] = 1
```

```python
#Label Part B dataset of 2019
Part_B_data_df_2019['Exclusion'] = 0

for index, row in tqdm(Part_B_data_df_2019.iterrows()):
    if row['Rndrng_NPI'] in list_of_2019_fraud_data:
        Part_B_data_df_2019.at[index, 'Exclusion'] = 1
```

```python
#Sum the number of fraud labels in the data of 2021 
labeled_number = sum(Part_B_data_df_2021['Exclusion'] == 1)
labeled_number
```
